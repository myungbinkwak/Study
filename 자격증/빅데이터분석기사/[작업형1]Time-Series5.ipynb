#### 주어진 데이터(basic2.csv)에서 새로운 컬럼(1일 이전 시차 컬럼)을 만들고, Events가 1이면서 Sales가 1000000이하인 조건에 맞는 새로운 컬럼 합을 구하시오
데이터셋 : basic2.csv

```python
import os
from os.path import join

import pandas as pd
```

```python
file_NM = join('data', 'basic2.csv')
df = pd.read_csv(file_NM)
df.head(3)
```

```python
df['before_PV'] = df['PV'].shift(1)
df
```

```python
df['before_PV'] = df['before_PV'].fillna(method='bfill')
df
```

```python
result = df[(df['Events'] == 1) & (df['Sales']<=1000000)]
result
```
