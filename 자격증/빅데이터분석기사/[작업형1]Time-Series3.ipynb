#### 주어진 데이터에서 2022년 월별 Sales 합계 중 가장 큰 금액과 2023년 월별 Sales 합계 중 가장 큰 금액의 차이를 절대값으로 구하시오.단 Events컬럼이 '1'인경우 80%의 Salse값만 반영함
(최종값은 소수점 반올림 후 정수 출력)

- 데이터셋 : basic2.csv

```python
import os
from os.path import join

import pandas as pd
import numpy as np
```

```python
file_NM = join('data', 'basic2.csv')
df = pd.read_csv(file_NM)
df.head(3)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df
```

```python
def sales(x):
    if x['Events'] == 1:
        x['Sales2'] = x['Sales']*0.8
    else:
        x['Sales2'] = x['Sales']
    return x

df = df.apply(lambda x: sales(x), axis=1)
df
```

```python
# 2022년 월별 Sales 합계 중 가장 큰 금액
# 단 Events컬럼이 '1'인경우 80%의 Salse값만 반영함
df2022 = df[df['Year'] == 2022]
df2022_max = df2022.groupby('Month')['Sales2'].sum().max()
df2022_max
```

```python
# 2023년 월별 Sales 합계 중 가장 큰 금액
df2023 = df[df['Year'] == 2023]
df2023_max = df2023.groupby('Month')['Sales2'].sum().max()
df2023_max
```

```python
# 가장 큰 금액의 차이를 절대값
print(int(abs(df2022_max - df2023_max)))
```
