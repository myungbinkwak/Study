#### 문제1
- 데이터셋(basic1.csv)의 'f5' 컬럼을 기준으로 상위 10개의 데이터를 구하고,
- 'f5'컬럼 10개 중 최소값으로 데이터를 대체한 후,
- 'age'컬럼에서 80 이상인 데이터의'f5 컬럼 평균값 구하기

```python
file_NM = join('data', 'basic1.csv')
df = pd.read_csv(file_NM)
df
```

```python
# 데이터셋(basic1.csv)의 'f5' 컬럼을 기준으로 상위 10개의 데이터

df = df.sort_values('f5', ascending = False)
df.head(10)
```

```python
# 'f5'컬럼 10개 중 최소값으로 데이터를 대체

min_f5 = df['f5'][:10].min()
print(min_f5)

df['f5'][:10] = min_f5
df.head(10)
```

```python
# 'age'컬럼에서 80 이상인 데이터의'f5 컬럼 평균값 구하기

df_old = df[df['age'] >= 80]
print(df_old)

df_old['f5'].mean()
```

#### 문제2
- 데이터셋(basic1.csv)의 앞에서 순서대로 70% 데이터만 활용해서,
- 'f1'컬럼 결측치를 중앙값으로 채우기 전후의 표준편차를 구하고
- 두 표준편차 차이 계산하기

```python
df = pd.read_csv(file_NM)
df.head()
```

```python
# 데이터셋(basic1.csv)의 앞에서 순서대로 70% 데이터만 활용

df_70 = df.iloc[:70]
df_30 = df.iloc[70:]
```

```python
# 'f1'컬럼 결측치를 중앙값으로 채우기 전후의 표준편차를 구하고

# 채우기전의 표준편차

before_df_70_std = df_70['f1'].std()
print(before_df_70_std)

# 중앙값으로 대체

after_df_70 = df_70['f1'].fillna(df_70['f1'].median())
df_70['f1'] = after_df_70

# 채운후의 표준편차
after_df_70_std = df_70['f1'].std()
print(after_df_70_std)
```

```python
# 두 표준편차 차이 계산
print(before_df_70_std - after_df_70_std)
```

#### 문제3
- 데이터셋(basic1.csv)의 'age'컬럼의 이상치를 더하시오!
- 단, 평균으로부터 '표준편차*1.5'를 벗어나는 영역을 이상치라고 판단함

```python
df = pd.read_csv(file_NM)
df.head()
```

```python
# 평균
mean_age = df['age'].mean()
mean_age
```

```python
# 표준편차
std_age = df['age'].std()*1.5
std_age
```

```python
# 이상치 정의
max_outlier = mean_age + std_age
min_outlier = mean_age - std_age
print(max_outlier, min_outlier)
```

```python
# 합
df[(df['age'] >= max_outlier)|(df['age'] <= min_outlier)]['age'].sum()
```
