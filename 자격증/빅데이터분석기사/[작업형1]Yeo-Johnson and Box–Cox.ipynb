#### 수치형 변수 변환하기
주어진 데이터에서 20세 이상인 데이터를 추출하고 'f1'컬럼을 결측치를 최빈값으로 채운 후, f1 컬럼의 여-존슨과 박스콕스 변환 값을 구하고, 두 값의 차이를 절대값으로 구한다음 모두 더해 소수점 둘째 자리까지 출력(반올림)하시오
- 데이터셋 : basic1.csv

```python
import os 
from os.path import join

import pandas as pd
import numpy as np
```

```python
file_Nm = join('data', 'basic1.csv')
df = pd.read_csv(file_Nm)
df
```

```python
# 20세 이상인 데이터 추출
df = df[df['age'] >= 20]
df.describe()
```

```python
# f1의 결측치를 최빈값으로 채움
df['f1'] = df['f1'].fillna(df['f1'].mode()[0])
df
```

```python
# Yeo-Johnson and Box–Cox
from sklearn.preprocessing import power_transform

df['YG'] = power_transform(df[['f1']])
df['BC'] = power_transform(df[['f1']], method='box-cox')
```

```python
round(sum(np.abs(df['YG'] - df['BC'])), 2)
```
